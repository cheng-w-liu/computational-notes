```python
import math
import torch
```

```python
import seaborn as sbn 
import matplotlib.pyplot as plt
import matplotlib
matplotlib.style.use('ggplot')
%matplotlib inline 
```

```python
d = 128
T = 500
```

```python
assert d % 2 == 0
```

```python
PE = torch.zeros(d, T)
```

```python
logW = torch.arange(0, d, 2) * (-math.log(10000.) / d)
W = torch.exp(logW)[:, None]
W.shape 
```

```python
time_steps = torch.arange(T)[None, :]
time_steps.shape 
```

```python
Wt = W * time_steps
Wt.shape 
```

```python
PE[0::2, :] = torch.sin(Wt)
PE[1::2, :] = torch.cos(Wt)
```

```python
ax = sbn.heatmap(PE)
ax.set_xlabel('Time', fontsize=15, labelpad=15)
ax.set_ylabel('Embedding dim.', fontsize=15, labelpad=15)
_ = ax.set_title(f'Positional Embedding, d={d}, T={T}', fontsize=15)
```
